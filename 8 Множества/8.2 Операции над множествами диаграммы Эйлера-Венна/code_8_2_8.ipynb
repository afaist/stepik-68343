# Задание 8.2.8
Множества A и B содержат 5 и 6 элементов соответственно, а множество A∩B –
2 элемента. Сколько элементов в множестве A∪B?

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
venn2(subsets=(3, 4, 2), set_labels=("A", "B"))
plt.show()
```

Объединённое множество содержит 9 элементов (3+2+4)
