# Задание 8.2.9
Каждый ученик в классе изучает английский или французский язык. Английский язык изучает 25 человек, французский – 27, а оба языка – 18. Сколько учащихся в классе?

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
```

```python
venn2(subsets=(7, 9,18), set_labels=("Английский", "Французский"))
plt.show()
```

- Только английский: 25 - 18 = 7
- Только французский: 27 - 18 = 9
- Всего учеников:

```python
7 + 18 + 9
```

Ответ: 34
